# 2章 パーセプトロン

## 2.3 パーセプトロンの実装

### 2.3.3 重みとバイアスによる実装

```julia
function AND(x1, x2)
    x = [x1, x2]
    w = [0.5, 0.5]
    b = -0.7
    tmp = dot(x, w) + b
    tmp <= 0.0 ? 0 : 1
end
```

```julia
[(x1, x2, AND(x1, x2)) for x2=[0,1] for x1=[0,1]]
```

```julia
function NAND(x1, x2)
    x = [x1, x2]
    w = [-0.5, -0.5]
    b = 0.7
    tmp = dot(x, w) + b
    tmp <= 0.0 ? 0 : 1
end
```

```julia
[(x1, x2, NAND(x1, x2)) for x2=[0,1] for x1=[0,1]]
```

```julia
function OR(x1, x2)
    x = [x1, x2]
    w = [0.5, 0.5]
    b = -0.2
    tmp = dot(x, w) + b
    tmp <= 0.0 ? 0 : 1
end
```

```julia
[(x1, x2, OR(x1, x2)) for x2=[0,1] for x1=[0,1]]
```

## 2.5 多層パーセプトロン

### 2.5.2 XORゲートの実装

```julia
function XOR(x1, x2)
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    AND(s1, s2)
end
```

```julia
[(x1, x2, XOR(x1, x2)) for x2=[0,1] for x1=[0,1]]
```
